```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('performance_measurements.csv', sep=';')

print(df.dtypes)

# Plot the speedup factor for each file size and thread count
plt.figure(figsize=(10, 6))

# Plot speedup factor for all data points
for file_size in df['File Size'].unique():
    plt.plot(df[df['File Size'] == file_size]['Thread Count'], df[df['File Size'] == file_size]['Speedup Factor'], label=file_size)
plt.xlabel('Thread Count')
plt.ylabel('Speedup Factor')
plt.title('Speedup Factor for 10k Data Points')
plt.legend()
plt.grid(True)
plt.show()



# Plot the time taken for each file size and thread count
plt.figure(figsize=(10, 6))

# Plot time taken for all data points
for file_size in df['File Size'].unique():
    plt.plot(df[df['File Size'] == file_size]['Thread Count'], df[df['File Size'] == file_size]['Execution Time (seconds)'], label=file_size)
plt.xlabel('Thread Count')
plt.ylabel('Time Taken')
plt.title('Time Taken for 10k Data Points')
plt.legend()
plt.grid(True)
plt.show()


# Display table of data to copy to word
df.to_clipboard()
```
