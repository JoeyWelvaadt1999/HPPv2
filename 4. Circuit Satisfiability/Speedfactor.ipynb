```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Results.csv', sep=';')

# Plot the speedup factor
plt.plot(df['Number of Processes'][1:], df['Speedup Factor'][1:], marker='o', linestyle='-')
plt.xlabel('Number of Processes')
plt.ylabel('Speedup Factor')
plt.xticks(np.arange(0, 257, 32))
plt.title('Speedup Factor for Circuit Satisfiability')
plt.show()

# Plot the time
plt.plot(df['Time (s)'], df['Number of Processes'], marker='o', linestyle='-')
plt.ylabel('Number of Processes')
plt.xlabel('Time (s)')
# Add steps 1, 2, 4, 8, 16, 32, 64, 128, 256 spread evenly
plt.yticks(np.arange(0, 257, 32))
plt.title('Time for Circuit Satisfiability')
plt.show()
```
